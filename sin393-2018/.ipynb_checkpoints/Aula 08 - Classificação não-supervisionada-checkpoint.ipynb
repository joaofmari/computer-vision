```python
import numpy as np
from scipy import spatial
from sklearn import datasets, cluster
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

%matplotlib inline

np.random.seed(1234)
```

## K-means

* O k-means inicia com N objetos, caracterizados em termos dos seus respectivos vetores de características e busca classifica-los em K classes.
    * O número de classes, K, deve ser conhecido *a priori*.
* O algoritmo requer K protótipos (ou sementes), $P_i$. 
* Cada centroide define uma área de influencia no espaço de características.

### Pseudo-código

* Obter os K protótipos iniciais e armazena-los em uma lista $P$;
* Enquanto instável:
    * Calcular todas as distâncias entre cada protótipo (ou média), $P_i$, e cada vetor de características:
        * Gera uma matriz de distâncias K x N, D;
    * Usar a matriz D para identificar as caraterísticas mais semelhantes ao protótipo $P_i$.
        * Armazenar esses pontos em uma lista $L_i$;
    * Obter como novos protótipos os centroides dos pontos de características armazenados em cada lista $L_i$.

## Implementação do início

```python
class Kmeans():
    
    def fit(self, X, P, plot=True):
        """
        
        """
        # N : Numero total de objetos
        # M : Numero de caracteristicas
        N, M = X.shape
        
        # Número de protótipos
        K = P.shape[0]
        
        # Critério de parada
        m = np.Inf
        m_min = 0.01
        
        colors = ['r', 'g', 'b', 'c', 'y', 'm']
        
        # Contador
        i=0
        while m > m_min:
            print('==== Iteracao ', i, ' ====')
            i = i+1
            
            # Calcula a matriz de distancias (D):
            D = spatial.distance.cdist(P, X, metric='euclidean')
            # TESTE
            print('Matriz de distancias(D):')
            print(np.array_str(D, precision=2))
            
            # Localiza os vetores protótipo mais próximos de cada ponto de dados.
            L = D.argmin(axis=0)
            # TESTE
            print('L: ')
            print(np.array_str(L, precision=2))
            
            # Armazena os valores dos vetores protótipos na ultima iteracao.
            P_old = P.copy()
            
            # Move (atualiza) os vetores de prototipo
            for l in range(K):
                P[l] = X[L==l,:].mean(axis=0)
            # TEST
            print('P_old:')
            print(np.array_str(P_old, precision=2))
            print('P: ')
            print(np.array_str(P, precision=2))
            
            if plot==True:
                plt.figure(figsize=(5,5))
                ## plt.subplot(1,2,2)
                for k in range(K):
                    plt.scatter(X[L==k][:,0], X[L==k][:,1], color=colors[k])
                    plt.scatter(P[k,0], P[k,1], color=colors[k], marker='^', s=64, edgecolors='k')
                    plt.scatter(P_old[k,0], P_old[k,1], color=colors[k], marker='s', s=64, edgecolors='k')

                    # plt.plot([x0_min, x0_max], [f_x0_min, f_x0_max], color='b')
                    plt.plot([P_old[k,0], P[k,0]], [P_old[k,1], P[k,1]], color='k')

                plt.xlabel('x_0')
                plt.ylabel('x_1')
                plt.title(str('Iteração ' + str(i-1)))
                plt.show()
            
            # Distancia entre a posicao dos vetores atual e anterior.
            m = np.max( np.sqrt( ((P_old - P)**2).sum(0) ) )
            # TESTE
            print('m: ')
            print(m)
    
        return [P, L]


X_treino = np.array([[0.5, 0.5],
                         [0.5, 2.5],
                         [1.0, 2.0],
                         [2.0, 3.0],
                         [1.5, 4.0],
                         [3.5, 4.5],
                         [3.5, 5.5],
                         [5.0, 5.5],
                         [4.5, 1.0],
                         [5.0, 2.0],
                         [5.0, 3.0],
                         [5.5, 2.0]])

P_treino = np.array([[0.5, 0.5],
                     [0.5, 1.0],
                     [1.0, 0.5]])


# Constrói o objeto k-means
kmeans = Kmeans()

# Treina o k-means.
p, pred = kmeans.fit(X_treino, P_treino)

print('\n=====FINAL:=====')
print(p)
print(pred)
```

### Teste com o conjunto de dados IRIS

```python
# Carrega o conjunto de dados
iris = datasets.load_iris()

# Conjunto de dados IRIS com todas as 3 classes
X = iris.data
y = iris.target

# Selecionado apenas duas caracteristicas:
# largura da sépala (1) e largura da pétala (3)
X = X[:, [1,3]]

y[y==1] = 0
y[y==3] = 1
```

```python
class Kmeans():
    
    def fit(self, X, P, plot=True):
        """
        
        """
        # N : Numero total de objetos
        # M : Numero de caracteristicas
        N, M = X.shape
        
        print(P)
        
        # Número de protótipos
        K = P.shape[0]
        
        # Critério de parada
        m = np.Inf
        m_min = 0.01
        
        colors = ['r', 'g', 'b', 'c', 'y', 'm']
        
        # Contador
        i=0
        while m > m_min:
            print('==== Iteracao ', i, ' ====')
            i = i+1
            
            # Calcula a matriz de distancias (D):
            D = spatial.distance.cdist(P, X, metric='euclidean')
            # TESTE
            print('Matriz de distancias(D):')
            print(np.array_str(D, precision=2))
            
            # Localiza os vetores protótipo mais próximos de cada ponto de dados.
            L = D.argmin(axis=0)
            # TESTE
            print('L: ')
            print(np.array_str(L, precision=2))
            
            # Armazena os valores dos vetores protótipos na ultima iteracao.
            P_old = P.copy()
            
            # Move (atualiza) os vetores de prototipo
            for l in range(K):
                P[l] = X[L==l,:].mean(axis=0)

            # TEST
            print('P_old:')
            print(np.array_str(P_old, precision=2))
            print('P: ')
            print(np.array_str(P, precision=2))
            
            if plot==True:
                plt.figure(figsize=(7,7))
                ## plt.subplot(1,2,2)
                for k in range(K):
                    plt.scatter(X[L==k][:,0], X[L==k][:,1], color=colors[k])
                    plt.scatter(P[k,0], P[k,1], color=colors[k], marker='^', s=64, edgecolors='k')
                    plt.scatter(P_old[k,0], P_old[k,1], color=colors[k], marker='s', s=64, edgecolors='k')

                    # plt.plot([x0_min, x0_max], [f_x0_min, f_x0_max], color='b')
                    plt.plot([P_old[k,0], P[k,0]], [P_old[k,1], P[k,1]], color='k')

                plt.xlabel('x_0')
                plt.ylabel('x_1')
                plt.title(str('Iteração ' + str(i-1)))
                plt.show()
            
            # Distancia entre a posicao dos vetores atual e anterior.
            m = np.max( np.sqrt( ((P_old - P)**2).sum(0) ) )
            # TESTE
            print('m: ')
            print(m)
    
        return [P, L]


# Constrói o objeto k-means
kmeans = Kmeans()

# Treina o k-means.
p, pred = kmeans.fit(X, np.array([[3, 1], [3.5, 1.5],[3, 1.5]]))

print('\n=====FINAL:=====')
print(p)
print(pred)
```

## Implementação Scikit-learn

```python
# SKLEARN Classificadores 
clf = cluster.KMeans(n_clusters=3)
# Etapa de Treinamento
clf.fit(X)
        
# Etapa de classificacao
classificacao = clf.labels_
P = clf.cluster_centers_
# TESTE
print(classificacao)
print(P)

map_a = ListedColormap(['r', 'g', 'b'])

plt.figure(figsize=(7,7))
## plt.subplot(1,2,2)

plt.scatter(X[:,0], X[:,1], c=classificacao, cmap=map_a)
plt.scatter(P[:,0], P[:,1], c=[0, 1, 2], cmap=map_a, marker='^', s=64, edgecolors='k')

plt.show()
```
