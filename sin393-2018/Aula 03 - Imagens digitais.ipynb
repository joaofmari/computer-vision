## Mostrando imagens na tela

A biblioteca matplotlib fornece uma série de métodos para plotar imagens na tela.

### Alterando o mapa de cores

Imagens em tons de cinza são mostradas, por padrão, usando um mapa de cores falso pelo matplotlib. É possível alterar o mapa de cores. Para mostrar uma imagem de intensidades com os respectivos níveis de cinza, pode-se utilizar o mapa de cores "gray".

```python
# Cria uma nova figura
plt.figure() 

# Plota a imagem na tela com a mapa de cores padrão do matplotlib
plt.imshow(im)
plt.title('Mapa de cores padrão do matplotlib')
```

```python
# Cria uma nova figura
plt.figure() 

# Plota a imagem na tela com a mapa de cores 'gray'
plt.imshow(im, cmap='gray')
plt.title('Mapa de cores alterado - gray')
```

### Plotando regiões de interesse

Como uma imagem é na verdade um numpy array, é possível selecionar pixels ou regiões de interesse usando indexação e fatiamento.

```python
# Cria uma nova figura
plt.figure() 

# Define uma região de interesse (RdI) usando fatiamento.
plt.imshow(im[200:220, 200:220], cmap='gray')
plt.title('RdI com 20x20 pixels')

# Mostra as figuras na tela
plt.show() 
```

Destancando a região de interesse selecionada, na imagem original.

```python
from matplotlib.patches import Rectangle

plt.figure() # Cria uma nova figura
plt.imshow(im2, cmap='gray')

axis = plt.gca() # Get the current axes
axis.add_patch(Rectangle((200,200), 20, 20, facecolor='none', edgecolor='red'))
plt.title('RdI destacado na imagem')
```

### Mostrando a barra de cores

### Plotando multiplas imagens em uma figura

O método subplot permite dividir a figura em várias áreas de plotagem.

Sintaxe: plt.subplot(num_lin, num_col, pos)

```python
plt.figure() 
# Divide a área de plotagem: 1 linha e duas colunas.
plt.subplot(1,2,1) # A área ativa é a 1.
plt.imshow(im)
plt.axis('off')
plt.title('Imagem sem os eixos')
plt.subplot(1,2,2) # Agora a área ativa é a 2.
plt.imshow(im, cmap='gray')
plt.title('Imagem (gray).')

plt.tight_layout()
plt.show()
```

### Imagens coloridas

```python
im_rgb = misc.face()
# im_r, im_g, im_b = im_rgb[:,:,0], im_rgb[:,:,1], im_rgb[:,:,2]
im_r = im_rgb[:,:,0]
im_g = im_rgb[:,:,1]
im_b = im_rgb[:,:,2]
    
plt.figure()
plt.subplot(2,2,1)
plt.imshow(im_rgb)
plt.title('Imagem RGB')
plt.subplot(2,2,2)
plt.imshow(im_r, cmap='Reds')
plt.title('Canal Vermelho')
plt.subplot(2,2,3)
plt.imshow(im_g, cmap='Greens')
plt.title('Canal Verde')
plt.subplot(2,2,4)
plt.imshow(im_b, cmap='Blues')
plt.title('Canal Azul')

plt.tight_layout()
plt.show()
```

### Plotando imagens e seus histogramas

```python
im_gray = misc.ascent()
im_rgb = misc.face()

plt.figure()
plt.subplot(2,2,1)
plt.imshow(im_gray, cmap='gray')
plt.subplot(2,2,2)
plt.hist(im_gray.flatten(), 256)
plt.subplot(2,2,3)
plt.imshow(im_rgb)
plt.subplot(2,2,4)
plt.hist(im_rgb[:,:,0].flatten(), 256, color='red', histtype='step')
plt.hist(im_rgb[:,:,1].flatten(), 256, color='green', histtype='step')
plt.hist(im_rgb[:,:,2].flatten(), 256, color='blue', histtype='step')

plt.tight_layout()
plt.show()
```

## Informações sobre as imagens

## Transformações geométricas

```python
import imageio
import matplotlib.pyplot as plt

im = imageio.imread('minha_imagem.png')

plt.figure()

plt.imshow(im, cmap='gray')

plt.show()
```

### Recorte (crop) usando fatiamento

```python
num_l, num_c = im.shape
(num_l, num_c)
```

```python
im_rec = im[num_l//4:-num_l//4, num_c//4:-num_c//4]

plt.figure()
plt.imshow(im_rec)

plt.show()
```

### Atribuição de valores usando fatiamento

```python
im_atr = im.copy() # Força a cópia da imagem

import numpy as np

im_atr[100:200, num_c//2:] = 255

plt.figure()
plt.imshow(im_atr, cmap='gray')
plt.show()
```

```python
im_atr[100:200, num_c//2:] = np.arange(num_c//2)

plt.figure()
plt.imshow(im_atr, cmap='gray')
plt.show()
```

### Espalhamento (reflexão) de uma imagem

a) Espelhamento vertical, ud -> up <--> down.

```python
import numpy as np

im_esp_v = np.flipud(im)

plt.figure()
plt.imshow(im_esp_v, cmap='gray')

plt.show()
```

```python
im_esp_h= np.fliplr(im)

plt.imshow(im_esp_h, cmap='gray')

plt.show()
```

### Rotação de imagens

a) Redimensiona o arranjo para conter a imagem rotacionada.

```python
from scipy import ndimage

im_rot = ndimage.rotate(im, 45)

plt.figure()
plt.imshow(im_rot, cmap='gray')

plt.show()
```

b) A imagem rotacionada é recortada para caber no arranjo.

```python
im_rot_r = ndimage.rotate(im, 135, reshape=False)

plt.figure()
plt.imshow(im_rot_r, cmap='gray')

plt.show()
```

## Exemplo: Aplicativo simples via linha de comando

Escrever o código a seguir em um arquivo .py e executar via linha de comando conforme instruções no DocString da função.

```python
import time
import numpy as np
from scipy import ndimage 
import imageio
import matplotlib.pyplot as plt
 
def main(dict_param):
    """
    Espelhamento ou rotação de uma imagem.
    
    Executa uma operação de espelhamento ou de rotação sobre uma imagem. 
    Plota o resultado na tela e salva o resultado em arquivo.
    Parameters 
    ----------
    filename_in: string
        Nome do arquivo de entrada.
    filename_out: string
        Nome do arquivo de saída.
    param_1: {string}
        Operação: 'esp' para espelhamento e 'rot' para rotação.
    param_2: {string, float}, opcional
        Se param_1 = 'esp': 'param_2' pode ser 'h' ou 'v'. 
        Se param_2 = 'rot': 'param_2' define o ângulo de rotação. 
    Notes
    -----
    Exemplo simples de um programa em Python que pode  ser executado via linha de comando (prompt).
    Examples
    --------
    C:\> python command_line.py lena.tif lena_out_esp.tif esp h
    C:\> python command_line.py lena.tif lena_out_rot.tif rot 90
    """
    # Os parametros são passados em um dicionário
    print('Dicionario de parametros: ' + str(dict_param))
    
    # Inicia a contagem de tempo
    t1 = time.clock()
    
    # Lê a imagem em filename_in
    im = imageio.imread(dict_param['filename_in'])

    # Escolhe  e executa as operações definidas nos argumentos
    if dict_param['param_1'] == 'esp':
        if dict_param['param_2'] == 'h':
            im_out = np.flipud(im)
        elif dict_param['param_2'] == 'v':
             im_out = np.fliplr(im)
        else:
            print('Parametro 2 invalido para operacao de espelhamento!')

    elif dict_param['param_1']=='rot':
            im_out = ndimage.rotate(im, float(dict_param['param_2']))

    else:
        print('Parametro 1 invalido!')
        return

    # Finaliza contagem de tempo.
    t2 = time.clock()

    # Plota as imagens em uma figura.
    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(im, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(im_out, cmap='gray')
    
    # Grava a imagem em arquivo.
    imageio.imsave(dict_param['filename_out'], im_out)

    # Contagem de tempo...
    timetaken = t2-t1
    print("Tempo = %f segundos" % timetaken)

    # Mostra as imagens na tela.
    plt.show()

if __name__ == '__main__':
    import sys
    print('Numero de argumentos: %i' % len(sys.argv))
    print('Lista de argumentos: ' + str(sys.argv))
    # Converte a lista sys.argv para um dicionário
    dict_param = {'filename_in'  : sys.argv[1],
                    'filename_out' : sys.argv[2],
                    'param_1'      : sys.argv[3],
                    'param_2'      : sys.argv[4]}

    # Chama a função main    
    main(dict_param)
```
